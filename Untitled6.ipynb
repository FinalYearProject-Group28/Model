```python
from google.colab import drive
from keras import models
from keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing import  image
import numpy as np
from keras.applications.vgg16 import preprocess_input
import matplotlib.pyplot as plt
import os
```

```python
model=models.load_model('/content/gdrive/MyDrive/FYP/covid_prediction.h5')
```

```python
model.summary()
```

```python
test_path="/content/Test"
```

```python
test_datagen = ImageDataGenerator(rescale = 1./255)
```

```python
test_set = test_datagen.flow_from_directory(
    directory = test_path,
    target_size = (224, 224),
    batch_size = 32,
    shuffle=False,
    class_mode = 'categorical'
)
```

```python
eva=model.evaluate(test_set)
```

```python
def model_predict(img,actual):
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis = 0)
    x_processed = preprocess_input(x)
    result = model.predict(x_processed)
    if(result[0][0]<.50):
        result="normal"
    else:
        result="covid"
        
    plt.figure()
    plt.imshow(img)
    plt.title('Actual : {} --> Predicted  : {}'.format(actual, result))
```
