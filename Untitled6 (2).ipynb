```python
labels={'covid': 0, 'normal': 1}
```

```python
def model_predict(i_path,actual):
    img=image.load_img(i_path,target_size = (224, 224))
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis = 0)
    x_processed = preprocess_input(x)
    result = model.predict(x_processed)
    predicted_class_indices=np.argmax(result,axis=1)
    labels2 = dict((v,k) for k,v in labels.items())
    predictions = [labels2[k] for k in predicted_class_indices]
    # if(result[0][0]<.50):
    #     result="normal"
    # else:
    #     result="covid"
    print(predictions)
    plt.figure()
    plt.imshow(img)
    plt.title('Actual : {} --> Predicted  : {}'.format(actual, predictions[0]))

    
```

```python
predict_path="/content/gdrive/MyDrive/FYP/predictions"
predict_datagen = ImageDataGenerator(rescale = 1./255)
predict_set = predict_datagen.flow_from_directory(
    directory = predict_path,
    target_size = (224, 224),
    batch_size = 32,
    shuffle=False,
    class_mode = 'categorical'
)
```

```python
label = predict_set.classes
```

```python
pred= model.predict(predict_set)
predicted_class_indices=np.argmax(pred,axis=1)
```

```python
labels2 = dict((v,k) for k,v in labels.items())
predictions = [labels2[k] for k in predicted_class_indices]
print(predictions)
```

```python
from sklearn.metrics import confusion_matrix

cfm = confusion_matrix(predicted_class_indices,label)
cfm
```

```python
plt.matshow(cfm)
plt.title('Confusion Matrix Plot')
plt.colorbar()
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show();
```
